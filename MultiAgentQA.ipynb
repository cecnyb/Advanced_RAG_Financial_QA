```python
from langchain_core.prompts import PromptTemplate

extract_ticker_prompt = PromptTemplate(
    input_variables=["question"],
    template="""
    You have to extract the company name in the given question, and answer the ticker of the company.
    Answer in JSON format, with 'ticker' key.

    Question: {question}

    Output JSON: {{
      "ticker": "<ticker of the company>"
    }}
    """
)

from pydantic import BaseModel, Field

class Ticker(BaseModel):
    """Ticker of the company in the given question"""
    ticker: str = Field(description="ticker of the company")

extract_ticker_chain = extract_ticker_prompt | llm.with_structured_output(Ticker)
```

```python
extract_ticker_chain.invoke({"question": "What was the revenue of Union Pacific Corp in 2022?"})
```

```python
from langchain_core.output_parsers import StrOutputParser
from datetime import datetime

parser = StrOutputParser()

change_prompt = PromptTemplate(
    input_variables=["original_question", "fiscal_info", "today"],
    template="""
    You have to convert relative time information in the given original question to the fiscal year of the company.
    Use the fiscal end date to convert to fiscal year.
    Answer only the converted question.

    Original question: {original_question}
    Fiscal end date for the company: {fiscal_info}
    Today date: {today}
    """
)

convert_chain = change_prompt | llm | parser

def get_fiscal_info_with_ticker(ticker: str) -> str:
    """Get fiscal year information of the company with ticker string."""
    ticker = ticker.upper()
    if ticker in company_fiscal_year_end:
      end_date = company_fiscal_year_end[ticker]
      return end_date[:2] + '-' + end_date[2:]
    else:
      return '12-31'

extract_fiscal_prompt = PromptTemplate(
    input_variables=["question"],
    template="""
    You have to extract the fiscal year in the given question.
    Answer in JSON format, with 'fy' key.

    Question: {question}

    Output JSON: {{
      "fy": <fiscal year in the question>
    }}
    """
)

from pydantic import BaseModel, Field

class FiscalYear(BaseModel):
    """FiscalYear mentioned in the given question"""
    fy: int = Field(description="fiscal year in the question")

extract_fiscal_chain = extract_fiscal_prompt | llm.with_structured_output(FiscalYear)
```

```python
from langchain_core.tools import tool

@tool
def rewrite2fiscal(question: str) -> tuple[str, int]:
  """Extract the company and convert relative time information in the question to the actual fiscal year of the company.
     It uses today function for the date information.

    Args:
        question: Question to be extracted.

    Returns:
        A tuple of ticker of the company and the fiscal year
  """
  ticker = extract_ticker_chain.invoke({"question": question}).ticker
  today = datetime.now().strftime("%Y-%m-%d")
  response = convert_chain.invoke({"original_question": question, "fiscal_info": get_fiscal_info_with_ticker(ticker), "today": today})
  fy = extract_fiscal_chain.invoke({'question': response}).fy
  return ticker, fy
```

```python
@tool
def calculate_eps(net_income: float, outstanding_shares: int):
  """Calculate the EPS of the company using net income and outstanding share

   Args:
        net_income: Net income value of the company
        outstanding_shares: Total stock held by the company's shareholders

    Returns:
        EPS value
  """
  if (net_income is None) or (outstanding_shares is None):
    print("Give me the net income and outstanding shares first")
    return None
  elif outstanding_shares == 0:
    print("Outstanding shares cannot be zero")
    return None

  return net_income / outstanding_shares
```

```python
@tool
def calculate_cashflowfromoperations(net_income: float, non_cash_items: float, changes_in_working_capital: float):
  """Calculate the cash flow from operations of the company using net income, non cash items and change in working capital

   Args:
        net_income: Net income value of the company
        non_cash_items: Financial transactions or events that are recorded in a company's financial statements but do not involve the exchange of cash
        changes_in_working_capital: Difference in a company's working capital between two reporting periods

    Returns:
        Value of cash flow from operations
  """
  if (net_income is None) or (non_cash_items is None) or (changes_in_working_capital is None):
    print("Give me the net income, non cash items and change in working capital data first")
    return None

  return net_income + non_cash_items + changes_in_working_capital
```

```python
import json

def count_rows_in_json(file_path):
  """Counts the number of rows (objects) in a JSON file."""
  try:
    with open(file_path, 'r') as f:
      data = json.load(f)
      if isinstance(data, list):
        return len(data)
      elif isinstance(data, dict):
        return len(data)
      else:
        return 0  # Or raise an error, depending on your needs
  except FileNotFoundError:
    print(f"File not found: {file_path}")
    return 0
  except json.JSONDecodeError:
    print(f"Error decoding JSON from file: {file_path}")
    return 0

# Assuming your file is named 'test.json' in the current directory
num_rows = count_rows_in_json('test.json')
print(f"Number of rows in the JSON file: {num_rows}")
```

```python
import json

def analyze_json(file_path):
  """Analyzes a JSON file, printing information about its attributes."""
  try:
    with open(file_path, 'r') as f:
      data = json.load(f)
      if isinstance(data, list):
        print("JSON file contains a list of objects.")
        if len(data) > 0:
          print("Example object:")
          print(json.dumps(data[0], indent=2))
          for key in data[0].keys():
            print(f"- Attribute: {key}")
      elif isinstance(data, dict):
        print("JSON file contains a dictionary.")
        print("Attributes:")
        for key in data.keys():
          print(f"- Attribute: {key}")
      else:
        print("JSON file is empty or contains an unexpected data type.")
  except FileNotFoundError:
    print(f"File not found: {file_path}")
  except json.JSONDecodeError:
    print(f"Error decoding JSON from file: {file_path}")

# Analyze 'test.json'
analyze_json('train.json')
```

```python
import json

def create_distinct_items(file_path):
  """Loads a JSON file and prints a sample of its contents."""
  distinct_items_with_context = {}

  try:
    with open(file_path, 'r') as f:

      data = json.load(f)


      for item in data:
        filename = item['filename']
        pre_text = item['pre_text']
        post_text = item['post_text']
        table = item['table']

        #table_as_string = json.dumps(table)
        #print("table", table_as_string)

        #print(table)

        distinct_items_with_context[filename] = {'pre_text': pre_text, 'post_text': post_text, 'table': table}

    # print(data[0].keys())
    return distinct_items_with_context


  except FileNotFoundError:
    print(f"File not found: {file_path}")

  except json.JSONDecodeError:
    print(f"Error decoding JSON from file: {file_path}")


distinct_items_with_context = create_distinct_items('test.json')
```

```python
def merge_text_list(text_list):

  return " ".join(text_list)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
import json

# Initialize text splitter and embeddings
splitter = RecursiveCharacterTextSplitter(chunk_size=1100, chunk_overlap=300)
embeddings = OpenAIEmbeddings(model='text-embedding-3-small', api_key=OPENAI_API_KEY)
chunks = []

# Function to process the table into text
def convert_table_to_text(table_str):
    """Convert a table JSON string into a readable text format."""
    if not table_str:  # Check if the table string is None or empty
        return "Table data is missing or null."

    try:
        # Attempt to parse the JSON string
        table_dict = json.loads(table_str)
    except json.JSONDecodeError:
        return "Table data is invalid JSON."

    # Extract unit and rows
    unit = table_dict.get("unit", "unknown unit")
    rows = []

    for row in table_dict.get("table", []):
        row_name = row["row_name"]
        values = ", ".join([f"{col}: {val}" for col, val in row["values"].items()])
        rows.append(f"Row: {row_name} | {values}")

    # Combine the unit and rows into a single text block
    return f"Unit: {unit}\n" + "\n".join(rows)

# Read the file containing tables
with open("/content/All_tables_structured.txt", "r") as file:
    all_lines = file.readlines()

# Assuming 'distinct_items_with_context' is a dictionary
i = 0
for filename, context_data in distinct_items_with_context.items():
    pre_text = merge_text_list(context_data['pre_text'])
    post_text = merge_text_list(context_data['post_text'])

    table = all_lines[i]
    columns = all_lines[i+1]
    table = table.replace("'", '"')
    table = table.replace("None", 'null')  # Fix JSON compatibility

    # Convert table to text
    table_text = convert_table_to_text(table)

    i += 2  # Increment to match the file structure

    # Split the pre-text and post-text into smaller chunks
    pre_text_chunks = splitter.split_text(pre_text)
    post_text_chunks = splitter.split_text(post_text)

    # Create a list of texts to store in the vector DB
    texts_for_db = []
    metadatas_for_db = []

    # Add pre-text chunks
    for chunk in pre_text_chunks:
        texts_for_db.append(chunk)
        metadatas_for_db.append({
            'company': filename.split('/')[0],
            'fiscal': int(filename.split('/')[1]),
            'context_type': 'pre_text'
        })

    # Add table text as a single chunk

    texts_for_db.append(table_text)
    metadatas_for_db.append({
        'company': filename.split('/')[0],
        'fiscal': int(filename.split('/')[1]),
        'column_names': columns,
        'context_type': 'table'
    })

    # Add post-text chunks
    for chunk in post_text_chunks:
        texts_for_db.append(chunk)
        metadatas_for_db.append({
            'company': filename.split('/')[0],
            'fiscal': int(filename.split('/')[1]),
            'context_type': 'post_text'
        })

    # Create the vector database for each file
    docsearch = Chroma.from_texts(
        texts_for_db,
        embeddings,
        metadatas=metadatas_for_db,
        persist_directory="./db"
    )
    chunks.append(docsearch)

```

```python
@tool
def calculate_operating_profit_margin(operating_profit, revenue):
    """
    Calculate the operating profit margin.

    Args:
    operating_profit (float): The operating profit of the company.
    revenue (float): The total revenue of the company.

    Returns:
    float: The operating profit margin as a percentage.
    """
    if revenue == 0:
        return 0
    print("op prof", operating_profit)
    print("rev", revenue)
    result =  (operating_profit / revenue) * 100
    print(result)
    return round(result, 8)

```

```python
@tool
def calculate_percentage_change(start_value: float, end_value: float) -> float:
    """
    Calculate the percentage change between two values.

    Args:
    start_value (float): The starting/beginning value.
    end_value (float): The ending value.

    Returns:
    float: The percentage change between the two values.
    """
    if start_value == 0:
      start_value = 0.00000001

    return round(((end_value - start_value) / start_value) * 100, 2)
```

```python
@tool
def classify_query(question: str) -> str:
    """
    Classify the query as low_level, medium_level or high_level.

    Args:
    question (str): The question to be classified.

    Returns:
    str: The difficulty level of the question.
    """

    system_query = """
Your task is to classify the financial query as low_level, medium_level or high_level. Query: {query}.

Use the following strategy:
1. See if the query is a low level query.
Low_level queries can be answered using retrieve_factual_data. Use the tool retrieve_factual_data to see if the nessecary context can be retrieved using simple RAG.
Alayze if the retrieved information can answer the query. If so, output low_level.
2. See if the query is a medium level query.
Medium level queries can be answered by first applying any of the relevant tools: calculate_eps, calculate_cashflowfromoperations, operating_profit_margin and percentage_change.
Then retrieve_factual_data can be used to retrieve the input for the tools or other relevant context. If so, output medium_level.
3. See if the query is a high level query.
High level queries need to use avdanced RAG to retreive the relevant context. If the nessecary context could not be retrived using retrieve_factual_data with or without tools, output high_level.

Important! Only output the difficulty level without any additional information.
Output: {{difficulty_level}}
    """
    tools = [
        rewrite2fiscal,
        retrieve_factual_data,
        calculate_eps,
        calculate_cashflowfromoperations,
        calculate_operating_profit_margin,
        calculate_percentage_change
    ]
    user_query = f"Question: {question}"
    print("user_query", user_query)

    # 3. Initial conversation
    messages = [
        SystemMessage(content=system_query),
        HumanMessage(content=user_query),
    ]
    llm_with_tools = llm.bind_tools(tools)

    # 4. Invoke the LLM in a loop, letting it call tools as needed
    ai_msg = llm_with_tools.invoke(messages)
    while True:
        # If AI provided a final answer and no new tool calls, we stop.
        if ai_msg.content.strip() != "" and not ai_msg.tool_calls:
            break

        total_length = sum(len(message.content) for message in messages)
        if total_length > 11000:
          print("had to break")
          break
        messages.append(ai_msg)

        # For each tool call, run the tool and append result.
        for tool_call in ai_msg.tool_calls:
            selected_tool = {
                "rewrite2fiscal": rewrite2fiscal,
                "calculate_eps": calculate_eps,
                "calculate_cashflowfromoperations": calculate_cashflowfromoperations,
                "retrieve_factual_data": retrieve_factual_data,
                "calculate_operating_profit_margin": calculate_operating_profit_margin,
                "calculate_percentage_change": calculate_percentage_change,
                "classify_query": classify_query
            }[tool_call["name"]]
            tool_msg = selected_tool.invoke(tool_call)
            messages.append(tool_msg)

        # Re-invoke the LLM with updated conversation
        ai_msg = llm_with_tools.invoke(messages)

    return ai_msg.content




```

# Query Decomposition

```python
from pydantic import BaseModel, Field

decomposer_prompt = PromptTemplate(
    input_variables=["query"],
    template="""
You are a helpful assistant that prepares queries that will be sent to a search component.
Sometimes, these queries are very complex.
Your job is to simplify complex queries into multiple queries that can be answered
in isolation to eachother.
Question: {query}

If the query is simple, then keep it as it is.
Examples
1. Query: Did Apple or American Tower Corp make more money last year?
   Decomposed Questions: [Question(question='How much profit did Apple make last year?', answer=None), Question(question='How much profit did American Tower Corp make last year?', answer=None)]
2. Query: What is the difference of Operating Profit Margin between Air Products and Chemicals in 2016 and printing papers business of International Paper Company in 2001?
   Decomposed Questions: [Question(question='What was the Operating Profit Margin Air Products and Chemicals in 2016?', answer=None)], [Question(question='What was the Operating Profit Margin printing papers business in 2001?', answer=None)]
3. Query: {{question}}
   Decomposed Questions:
"""
)

class Question(BaseModel):
    question: str = Field(description="question to be answered")


class Questions(BaseModel):
    questions: list[Question] = Field(description="questions to be answered")


decomposer_chain = decomposer_prompt | llm.with_structured_output(Questions)
```

```python
decomposer_chain.invoke({"query": "What is the difference of Union Pacific Corp's revenue from 2013 to 2016 divided the sum of Operating Profit Margin between Air Products and Chemicals in 2015 and printing papers business of International Paper Company in 2004?"})
```

```python
@tool
def decompose_query(query:str):
  '''
   Decomposes a query that conists of subquestions.

    Args:
    query (str): The question to be decomposed.

    Returns:
    list[Questions]: The list of decomposed questions.
  '''
  result = decomposer_chain.invoke({"query": query}).questions
  print(result)
  return result
```

```python
@tool
def advanced_retrieval(query: str) -> str:
  """
  Uses advanced methods to retrieve context related documents

  Args:
  query (str): The question to be answered.

  Returns:
  str: The retrieved documents.
  """


  system_query = """
Your task is to retrieve important contextual documents to answer the financial query.
The queries are a bit tricky, so you have to use smart techniques to find the relevant documents.

Query: {query}.

Use the following strategy:
1. Filter out keywords in the query and use synonyms.
Example:
   Query: Air Products current ratio 2024
   Keywords:
       - current ratio
       - liquidity ratio
       - cash flow ratio
       - working capital ratio
       - liquidity
Also try to keep the query short when retrieving documents.
Example 2:
Query: What was the Operating Profit Margin of the printing papers business of International Paper Company in 2005?
Shortened query: Profit printing International Paper Company 2005

3. Conduct the search by using the search tools provided: hybrid_search or retrieve_factual_data.
If you find a relevant unit in the query, or have a guess on the relevant unit for the answer, you can search on this using retrieve_factual_data.

Example query with unit:
Query: What was the Operating Profit Margin of the printing papers business of International Paper Company in 2005?
Unit query: Paper Company 2005 million dollars $

Additionally, you can also try using the previous or next year instead.

4. Try different approaches until you receive the context you need.

5. When you have found the context, retrieve the relevant documents.

For your help, you can call tools.
Output: {{Relevant context}}
"""
  tools = [
      rewrite2fiscal,
      hybrid_search,
      retrieve_factual_data
  ]
  user_query = f"Question: {query}"
  print("user_query", user_query)

  messages = [
      SystemMessage(content=system_query),
      HumanMessage(content=user_query),
  ]
  llm_with_tools = llm.bind_tools(tools)

  ai_msg = llm_with_tools.invoke(messages)
  while True:

      if ai_msg.content.strip() != "" and not ai_msg.tool_calls:
          break

      total_length = sum(len(message.content) for message in messages)
      if total_length > 11000:
        messages = messages[500:]


      messages.append(ai_msg)

      for tool_call in ai_msg.tool_calls:
          selected_tool = {
              "rewrite2fiscal": rewrite2fiscal,
              "hybrid_search": hybrid_search,
              "retrieve_factual_data": retrieve_factual_data,
          }[tool_call["name"]]
          tool_msg = selected_tool.invoke(tool_call)
          messages.append(tool_msg)

      ai_msg = llm_with_tools.invoke(messages)

  return ai_msg.content


```

```python
@tool
def calculate_ratio(value1, value2):
  """
  Calculates the ratio between two values.

  Args:
  value1 (float): The first value.
  value2 (float): The second value.

  Returns:
  float: The ratio between the two values.
  """
  if value2 == 0:
    value2 = 0.00000001

  return abs(value1 / value2)
```

```python
@tool
def calculate_difference(value1, value2):
  """
  Calculates the difference between two values.

  Args:
  value1 (float): The first value.
  value2 (float): The second value.

  Returns:
  float: The difference between the two values.
  """
  return value1 - value2
```

```python
def agentic_rag(query: str) -> str:
    # 1. Tools
    tools = [
        decompose_query,
        rewrite2fiscal,
        classify_query,
        calculate_eps,
        calculate_cashflowfromoperations,
        calculate_operating_profit_margin,
        calculate_percentage_change,
        retrieve_factual_data,
        advanced_retrieval,
        calculate_ratio,
        calculate_difference
    ]

    system_query = """
You are a professional agent that can use tools to answer a financial question. To answer the questions, you can retrieve documents.
The documents consist of text and tables. The tables have row names and values; look for relevant row names to answer the query.

Use the following strategy to answer the question:
1. Decompose the query using the decompose_query tool.
2. For each subquery in the decomposed query:
  1. Classify the subquery using classify_query.
  If the subquery is low_level:
    Use retrieve_factual_data to get context and answer the subquery.
  Elif the subquery is medium_level:
    Use relevant tools. Use retrieve_factual_data to get the necessary context. Then answer the subquery.
  Elif the subquery is high_level:
    Use relevant tools if needed. Use advanced_retrieval to get the necessary context. Then answer the subquery.

3. Use the answers to the subquestions to answer the original question. Provide a concise but complete answer.
If you struggle with finding information, you may take help from the advanced_retrieval even for low or medium level questions.
    """

    user_query = f"Question: {query}"

    # 3. Initial conversation
    messages = [
        SystemMessage(content=system_query),
        HumanMessage(content=user_query),
    ]

    llm_with_tools = llm.bind_tools(tools)

    # 4. Invoke the LLM in a loop, letting it call tools as needed
    ai_msg = llm_with_tools.invoke(messages)
    i = 0
    while True:
        # If AI provided a final answer and no new tool calls, we stop.
        if ai_msg.content.strip() != "" and not ai_msg.tool_calls:
            break

        total_length = sum(len(message.content) for message in messages)
        if total_length > 11000:
          messages = messages[500:]

        messages.append(ai_msg)

        # For each tool call, run the tool and append result.
        for tool_call in ai_msg.tool_calls:
            selected_tool = {
                "rewrite2fiscal": rewrite2fiscal,
                "calculate_eps": calculate_eps,
                "calculate_cashflowfromoperations": calculate_cashflowfromoperations,
                "retrieve_factual_data": retrieve_factual_data,
                "calculate_operating_profit_margin": calculate_operating_profit_margin,
                "calculate_percentage_change": calculate_percentage_change,
                "classify_query": classify_query,
                "decompose_query": decompose_query,
                "advanced_retrieval": advanced_retrieval,
                "calculate_ratio": calculate_ratio,
                "calculate_difference": calculate_difference
            }[tool_call["name"]]
            tool_msg = selected_tool.invoke(tool_call)
            messages.append(tool_msg)

        # Re-invoke the LLM with updated conversation
        i += 1
        if i > 25:
          break
        ai_msg = llm_with_tools.invoke(messages)

    return ai_msg.content
```
